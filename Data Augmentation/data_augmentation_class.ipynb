```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from tqdm import tqdm
import cv2
import os
from tensorflow.keras.utils import array_to_img, img_to_array, load_img
```

```python
Data_path="F:\CVPR Dataset\Imagedata"
Data_path2="F:\CVPR Dataset\Augmented Data"
```

```python
if not os.path.exists(Data_path2):
    os.makedirs(Data_path2)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    brightness_range=[0.5,1.2]
)
```

```python
classes = os.listdir(Data_path)

for i, c in enumerate(classes):
    path = os.path.join(Data_path, c)
    path2 = os.path.join(Data_path2, c)

    # Create the target directory for the current class if it doesn't exist
    if not os.path.exists(path2):
        os.makedirs(path2)

    for img in tqdm(os.listdir(path)):
        img_arr = cv2.imread(os.path.join(path, img))
        
        # Convert image to RGB
        img_arr_rgb = cv2.cvtColor(img_arr, cv2.COLOR_BGR2RGB)
        
        x = img_to_array(img_arr_rgb)
        x = x.reshape((1,) + x.shape)
        j = 0

        # Perform data augmentation and save 50 different images
        for batch in datagen.flow(x, batch_size=1, save_to_dir=path2, save_prefix=c, save_format='jpeg'):
            j += 1
            if j > 50:
                break
```
