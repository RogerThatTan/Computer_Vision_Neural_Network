```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from tqdm import tqdm
import cv2
```

```python
img = Image.open('F:\CVPR Dataset\MOHAMMED TANVIR HASSAN\IMG_20204.jpg')
img = img.resize((256, 256))  # Resize the image if needed
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='nearest',
)
```

```python
img_array = np.array(img)
```

```python
input_batch = img_array.reshape(1,256,256,3)
```

```python
i = 0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='F:\CVPR Dataset\MOHAMMED TANVIR HASSAN'):
    i=i+1

    if i == 10:
        break
```
